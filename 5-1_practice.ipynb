```python
import pandas as pd
import seaborn as sns

titanic = sns.load_dataset("titanic")
titanic.head()
```

```python
titanic.columns
```

```python
data = titanic[['pclass', 'sex', 'age', 'sibsp', 'parch', 'fare', 'embarked']]
target = titanic['survived']
data_encoded = pd.get_dummies(data, drop_first=True)

from sklearn.model_selection import train_test_split
train_input, test_input, train_target, test_target = train_test_split(data_encoded,target,test_size=0.2,random_state=42)

from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(max_depth=6,min_samples_leaf=7,random_state=42)
dt.fit(train_input, train_target)
print(dt.score(train_input, train_target))
print(dt.score(test_input, test_target))
```

```python
from sklearn.impute import SimpleImputer
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LogisticRegression

# 1) 파이프라인 만들기 (Imputer + LogisticRegression)
model = make_pipeline(
    SimpleImputer(strategy="median"),  # NaN → 중앙값으로 채우기
    LogisticRegression(max_iter=1000)
)

# 2) 이 model로 학습
model.fit(train_input, train_target)

# 3) score도 model 기준으로
print(model.score(train_input, train_target))
print(model.score(test_input, test_target))
```
